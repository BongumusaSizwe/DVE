```python
import numpy as np
from matplotlib import pyplot as plt
from skimage import io
import warnings
warnings.filterwarnings("ignore")
```

# **1.3.1:**
- The projections of ISOMAP did not change with the change in number of components.
- The projections of UMAP changed as the number of components changed. The appear to lose the u-shape and became thinner and condensed as the number of components increased.
- The projections of Spectral Embedding did not change with the change in number of components.
- The LLE projections changed as the number of components changed. The change is in these projections were as follows: the projections first formed a straight line with a negative gradient, then they proceeded to form an L-shape like shape and became a single dot.

# **1.3.2:**
The ISOMAP and Spectral Embedding projections were more stable as the number of components increased.

# **1.3.3**

## **Isomap:**
The following plots show the different Isomap projections produced for different component sizes specifically, 20, 40, 60, 80, 100 components. The number of neighbours were kept constant at 100. 

```python
fig,axes = plt.subplots(nrows=3, ncols=2, figsize=(8,8), sharex=True, sharey=True)
ax=axes.ravel()


ax[0].imshow(io.imread("1.3 plots/ISOMAP - 20 components.png"), cmap=plt.cm.gray)

ax[1].imshow(io.imread("1.3 plots/ISOMAP - 40 components.png"), cmap=plt.cm.gray)


ax[2].imshow(io.imread("1.3 plots/ISOMAP - 60 components.png"), cmap=plt.cm.gray)

ax[3].imshow(io.imread("1.3 plots/ISOMAP - 80 components.png"), cmap=plt.cm.gray)

ax[4].imshow(io.imread("1.3 plots/ISOMAP - 100 components.png"), cmap=plt.cm.gray)

for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```
